# 02. Exploratory Data Analysis (EDA)
This notebook generates visualizations for distributions, correlations, and feature comparisons.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src import config
from src.preprocessing import load_raw_data, clean_data

df = clean_data(load_raw_data(config.RAW_DATA_PATH))
df.describe()
```

```python
# Disease Distribution
sns.countplot(x='target', data=df)
plt.title('Disease Distribution (0 = No Disease, 1 = Disease)')
plt.show()
```

```python
# Correlation Matrix Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
